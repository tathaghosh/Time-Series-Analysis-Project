Test Stationarity


```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value 
        print(dfoutput)
```

```python
#Checking Stationarity of the location 43314

test_stationarity(d1["rainfall"])
```

```python
#Checking Stationarity of the overall Data
frames=[a12,a22,a32,a42,a52,a62,a72,a82,a92]
result=pd.concat(frames)
```

```python
test_stationarity(result)
```

Eliminating and estimting trend

```python
ts_sqrt1=np.sqrt(a12)
plt.plot(ts_sqrt1)
```

```python
ts_sqrt1.tail()
```

```python
moving_avg = ts_sqrt1.rolling(12).mean()
plt.plot(ts_sqrt1)
plt.plot(moving_avg, color='red')
```

```python
ts_sqrt_moving_avg_diff = ts_sqrt1 - moving_avg
ts_sqrt_moving_avg_diff.head(12)
```

```python
ts_sqrt_moving_avg_diff.dropna(inplace=True)
test_stationarity(ts_sqrt_moving_avg_diff)
```

```python
#Converting ts_sqrt1 to Dataframe
df_sqrt1 = pd.DataFrame(ts_sqrt1)
df_sqrt1
```

```python
#Exponentially weighted moving average
expwighted_avg = df_sqrt1.ewm(halflife=12).mean()
plt.plot(ts_sqrt1)
plt.plot(expwighted_avg, color='red')
```

```python
ts_sqrt_ewma_diff = df_sqrt1 - expwighted_avg
ts_sqrt_ewma_diff.tail()
```

```python
test_stationarity(ts_sqrt_ewma_diff)
```

```python
#Eliminating Trend and Seasonality
ts_sqrt_diff = ts_sqrt1 - ts_sqrt1.shift()
plt.plot(ts_sqrt_diff)
```

```python
ts_sqrt_diff.dropna(inplace=True)
test_stationarity(ts_sqrt_diff)
```

```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(ts_sqrt_diff, nlags=20)
lag_pacf = pacf(ts_sqrt_diff, nlags=20, method='ols')
#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_sqrt_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_sqrt_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
#Plot PACF:
plt.subplot(122)
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_sqrt_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_sqrt_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

#p=1 and q=2 
Target is to estimate best parameter estimation

```python
#AR Model
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(ts_sqrt1, order=(2, 1, 0))  
results_AR = model.fit(disp=-1)  
plt.plot(ts_sqrt_diff)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-ts_sqrt_diff)**2))
```

```python
#MA Model
model = ARIMA(ts_sqrt1, order=(0, 1, 2))  
results_MA = model.fit(disp=-1)  
plt.plot(ts_sqrt_diff)
plt.plot(results_MA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_MA.fittedvalues-ts_sqrt_diff)**2))
```

```python
#Combined Model
model = ARIMA(ts_sqrt1, order=(2, 1, 2))  
results_ARIMA = model.fit(disp=-1)  
plt.plot(ts_sqrt_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-ts_sqrt_diff)**2))
```

```python
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
print (predictions_ARIMA_diff.head())
```

```python
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
print (predictions_ARIMA_diff_cumsum.head())
```

```python
predictions_ARIMA_sqrt = pd.Series(ts_sqrt1.ix[0], index=ts_sqrt1.index)
predictions_ARIMA_sqrt = predictions_ARIMA_sqrt.add(predictions_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA_sqrt.head()
```

```python
predictions_ARIMA = np.exp(predictions_ARIMA_sqrt)
plt.plot(a12)
plt.plot(predictions_ARIMA)
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_ARIMA-a12)**2)/len(a12)))
```
